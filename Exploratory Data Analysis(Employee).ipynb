```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
## Summary statistics for numerical fields

df.corr()
```

```python
# Check for Null values

df.isnull().sum()
```

```python
df.drop(columns ='EmpNumber',inplace = True) # Employee Number is not useful for us.
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
l = LabelEncoder()
```

```python
df.Gender=l.fit_transform(df.Gender)
df.EducationBackground=l.fit_transform(df.EducationBackground)
df.MaritalStatus=l.fit_transform(df.MaritalStatus)
df.EmpDepartment=l.fit_transform(df.EmpDepartment)
df.EmpJobRole=l.fit_transform(df.EmpJobRole)
df.BusinessTravelFrequency=l.fit_transform(df.BusinessTravelFrequency)
df.Attrition=l.fit_transform(df.Attrition)
df.OverTime=l.fit_transform(df.OverTime)
```

```python
import seaborn as sb
from matplotlib import rcParams
rcParams['figure.figsize'] = 10,5
sb.set_style('whitegrid')
sb.heatmap(df)
```
